```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfD = pd.read_csv('22deDiciembreDatilli.txt', header=0, skiprows=3, sep='\s+')
dfJ = pd.read_csv('22deJunioDatilli.txt', header=0, skiprows=3, sep='\s+')
```

```python
dfD=dfD.drop([0, 1,2])
dfJ=dfJ.drop([0, 1,2])
```

```python
dfD=dfD.drop(dfD.index[len(dfD)-1])
dfJ=dfJ.drop(dfJ.index[len(dfJ)-1])
```

```python
dfD=dfD.reset_index()
dfJ=dfJ.reset_index()
```

```python
dfD=dfD.drop('index', axis=1)
dfJ=dfJ.drop('index', axis=1)
```

```python
dfD=pd.DataFrame(dfD)
dfJ=pd.DataFrame(dfJ)
```

```python
dfD.dtypes
dfJ.dtypes
```

```python
cols=['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
dfD[cols]=dfD[cols].apply(pd.to_numeric, errors='ignore' ,axis=1)
dfJ[cols]=dfJ[cols].apply(pd.to_numeric, errors='ignore' ,axis=1)
```

```python
dfD.dtypes
dfJ.dtypes
```

```python
# Gráfica de la Presión con respecto a la Altura (Diciembre).
dfD1 = dfD[['HGHT','PRES']]
plt.figure(); dfD1.plot(x='PRES'); plt.legend(loc='best')
plt.title("Variación de la presión contra la altura del mes de Diciembre")
plt.ylabel("Altura(m)")
plt.gca().invert_xaxis()
plt.xlabel("Presión(hPa)")
plt.grid(True)
plt.show()
```

```python
# Gráfica de la Presión con respecto a la Altura (Junio).
dfJ1 = dfJ[['HGHT','PRES']]
plt.figure(); dfJ1.plot(x='PRES'); plt.legend(loc='best')
plt.title("Variación de la presión contra la altura del mes de Junio")
plt.ylabel("Altura(m)")
plt.gca().invert_xaxis()
plt.xlabel("Presión(hPa)")
plt.grid(True)
plt.show()
```

```python
# Gráfica de temperatura como función de la altura (Diciembre).
dfD1 = dfD[['TEMP','HGHT']]
plt.figure(); dfD1.plot(x='TEMP'); plt.legend(loc='best')
plt.title("Variación de la temperatura como función de la altura del mes de Diciembre")
plt.ylabel("Altura(m)")
plt.xlabel("Temperatura(C°)")
plt.xlim([-70,10])
plt.grid(True)
plt.show()
```

```python
# Gráfica de temperatura como función de la altura (Junio).
dfJ1 = dfJ[['TEMP','HGHT']]
plt.figure(); dfJ1.plot(x='TEMP'); plt.legend(loc='best')
plt.title("Variación de la temperatura como función de la altura del mes de Junio")
plt.ylabel("Altura(m)")
plt.xlabel("Temperatura(C°)")
plt.xlim([-70,10])
plt.grid(True)
plt.show()
```

```python
# Gráfica de la temperatura (TEMP) y temperatura de rocío (DWPT), como función de la altura
# (Diciembre).
dfD1 = dfD[['TEMP','DWPT']]
dfD2 = dfD[['HGHT']]
plt.title("Varación de la temperatura y la temperatura de rocio como función de la altura del mes de Diciembre")
plt.ylabel("Altura(m)")
plt.grid(True)
plt.plot(dfD1,dfD2)
plt.show()
```

```python
# Gráfica de la temperatura (TEMP) y temperatura de rocío (DWPT), como función de la altura
# (Diciembre).
dfJ1 = dfJ[['TEMP','DWPT']]
dfJ2 = dfJ[['HGHT']]
plt.title("Varación de la temperatura y la temperatura de rocio como función de la altura del mes de Junio")
plt.ylabel("Altura(m)")
plt.grid(True)
plt.plot(dfJ1,dfJ2)
plt.show()
```

```python
# Gráfica de la rapidez de los vientos en nudos (SKNT) (Diciembre).
plt.figure(); dfD.SKNT.plot(); plt.legend(loc='best')
plt.title("Variación de la Rapidez de los Vientos del mes de Diciembre")
plt.ylabel("Rapidez (nudos)")
plt.grid(True)
plt.show()
```

```python
# Gráfica de la rapidez de los vientos en nudos (SKNT) (Junio).
plt.figure(); dfJ.SKNT.plot(); plt.legend(loc='best')
plt.title("Variación de la Rapidez de los Vientos del mes de Junio")
plt.ylabel("Rapidez (nudos)")
plt.grid(True)
plt.show()
```

```python
# Gráfica de la humedad relativa como función de la altura (Diciembre).
dfD1 = dfD[['RELH','HGHT']]
plt.figure(); dfD1.plot(x='RELH'); plt.legend(loc='best')
plt.title("Variación de la humedad relativa como función de la altura del mes de Diciembre")
plt.ylabel("Altura(m)")
plt.xlabel("Porcentaje de humedad")
plt.xlim([0,100])
plt.grid(True)
plt.show()
```

```python
# Gráfica de la humedad relativa como función de la altura (Junio).
dfJ1 = dfJ[['RELH','HGHT']]
plt.figure(); dfJ1.plot(x='RELH'); plt.legend(loc='best')
plt.title("Variación de la humedad relativa como función de la altura del mes de Junio")
plt.ylabel("Altura(m)")
plt.xlabel("Porcentaje de humedad")
plt.xlim([0,100])
plt.grid(True)
plt.show()
```
