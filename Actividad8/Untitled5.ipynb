```python
aa = 1.0
bb = 2.0

def d2X_dt2(X, t=0):
    """ Return the Jacobian matrix evaluated in X. """
    return array([[ 0.,   1. ],
                  [ -u*bb*X[0]*X[1] -aa,   u -X[0]**bb] ])
```
